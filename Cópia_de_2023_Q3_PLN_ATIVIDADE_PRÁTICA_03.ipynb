#### **INTRODUÇÃO**

**Descrição da API**

A **Telegram Bot API** é uma API que permite que usuários criem e interajam com bots dentro da plataforma do Telegram. Os bots podem ser utilizados para diversas finalidades, incluindo:

*   **Substituir websites:** os bots do Telegram suportam aplicações escritas em *JavaScript*, criando uma interface tão versátil quanto a de um website;
*   **Receber pagamentos:** os bots permitem a integração com mais de 20 serviços de pagamento, oferecendo suporte à transações financeiras de forma segura e isenta de custos;
*   **Criar ferramentas personalizadas:** os bots podem ser programados para realizar diferentes tarefas de forma automatizada, como conversão de arquivos, gerenciamento de conversas e resumo da previsão do tempo;
*   **Integrar com outros serviços:** diversos serviços possuem bots do Telegram que podem ser intergados, como Gmail, GitHub, Bing, YouTube e Wiki;
*   **Hospedar jogos:** os bots oferecem suporte à jogos single player ou multiplayer escritos em HTML5;
*   **Construir redes sociais:** os bots oferecem uma plataforma de integração entre os usuários, conectando pessoas com base em interesses pessoais, localização, entre outros.

```
# Importando as bibliotecas necessárias:
import logging
from telegram import Update
from telegram.ext import ApplicationBuilder, ContextTypes, CommandHandler
```

#### **UTILIZAÇÃO BÁSICA**

É recomendado configurar um logger para registrar possíveis erros de execução.

```
# Configurando o logger:
logging.basicConfig(
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    level=logging.INFO
)
```

Agora, implementamos uma função que será chamada toda vez que o comando
/start for acionado no chat, indicando o início de uma conversa.

```
# Ao receber o comando /start, o bot responde "I'm a bot, please talk to me!"
async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await context.bot.send_message(
        chat_id=update.effective_chat.id,
        text="I'm a bot, please talk to me!"
    )
```

Como se pode ver, essa função recebe dois parâmetros:

*   **update:** objeto que contém todas as informações e dados que estão vindo do próprio Telegram (como a mensagem, o usuário que enviou o comando, etc.)
*   **context:** objeto que contém informações e dados sobre o estado da biblioteca em si (como o Bot, a Aplicação, a job_queue, etc).

```
# Configuração inicial do bot
if __name__ == '__main__':

    # Substituir 'TOKEN' pelo token fornecido pelo BotFather
    application = ApplicationBuilder().token('TOKEN').build()

    # Configurar um Handler para responder ao comando /start
    start_handler = CommandHandler('start', start)
    application.add_handler(start_handler)

    # Roda o bot até que as teclas CTRL+C sejam acionadas
    application.run_polling()
```

```
# Bibliotecas necessárias para a utilização

from typing import Final
from telegram import Update
from telegram.ext import *
import logging
import responses
import nest_asyncio
```

```
# Primeiro, coloque o token dado pelo BotFather
TOKEN = 'COLOQUE SEU TOKEN AQUI'

# Criando o logging

logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logging.info('Inicializando o bot...')

async def start_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
  await update.message.reply_text('Olá! Sou um robô!')

async def help_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
  await update.message.reply_text('Em que posso te ajudar?')

async def custom_command(update: Update, context: ContextTypes.DEFAULT_TYPE):
  await update.message.reply_text('Esse é um comando personalizado.')

async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
  text = str(update.message.text).lower()
  logging.info(f'Usuário ({update.message.chat.id}) disse: {text}')

  # Resposta do bot (é o que o próprio usuário disse)

  await update.message.reply_text(text)


async def error(update, context):
  logging.error(f'Update ({update}) caused error: {context.error}')

def main():

  # Cria uma aplicação com o token a ser utilizado

  application = Application.builder().token(TOKEN).build()


  # Comandos

  application.add_handler(CommandHandler('start', start_command))
  application.add_handler(CommandHandler('help', help_command))
  application.add_handler(CommandHandler('custom', custom_command))

  # Mensagens

  application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))

  # Rodando o bot

  # Precisamos utilizar o nest_asyncio para poder rodar o bot no Google Colab

  nest_asyncio.apply()

  application.run_polling(allowed_updates=Update.ALL_TYPES)

if __name__ == "__main__":
    main()
```

***O erro de RuntimeError aparece quando apertamos 'stop' para parar de compilar o código. Contudo, ao compilar o código, o bot continuará funcionando até apertarmos 'stop'.***
